```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "OrderID": [1,2,3,4,5],
    "Product": ["Shirt", "Phone", "Shoes", "Laptop", "Watch"],
    "Category": ["Fashion", "Electronics", "Fashion", "Electronics", "Accessories"],
    "Price": [500, 15000, 1200, 55000, 2000],
    "Quantity": [2,1,1,1,2],
    "Date": ["2025-01-10","2025-01-11","2025-01-12","2025-01-13","2025-01-14"]
}

df = pd.DataFrame(data)
df
```

```python
df["Total_Sales"] = df["Price"] * df["Quantity"]
df
```

```python
df["Total_Sales"].sum()
```

```python
df.groupby("Product")["Total_Sales"].sum().sort_values(ascending=False)
```

```python
df.groupby("Category")["Total_Sales"].sum().plot(kind="bar")

plt.title("Category Wise Sales")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"])

df.groupby("Date")["Total_Sales"].sum().plot(kind="line", marker="o")

plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
df.groupby("Category")["Total_Sales"].sum().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Category Share")
plt.ylabel("")
plt.show()
```
